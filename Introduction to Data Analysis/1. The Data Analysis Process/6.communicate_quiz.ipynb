```python
# imports and load data
import pandas as pd
% matplotlib inline

df = pd.read_csv('store_data.csv')
df.head()
```

```python
# Which store has the highest total sales for the last month?
last_month = df[df['week'] == df['week'].max()]
last_month.plot(kind='bar');
```

```python
# Which store makes the most sales on average?
df.mean().plot(kind='bar');
```

```python
# Which store sells the most during the week of March 13th, 2016?
most_sales = df[df['week'] == '2016-03-13']
most_sales.plot(kind='bar');
```

```python
# Which store has the most sales in the latest 3-month period?
three_months_sales = df.sort_values(by='week', ascending=False).head(3)
three_months_sales.iloc[:,1:].sum().plot(kind='bar');
```
